````python
# 2. Tampilkan image dalam channel red-blue dan green blue
img6 = cv.cvtColor(img, cv.COLOR_BGR2RGB)

# 0 = Red, 1` = Green, 2 = Blue

# Red-Blue channel, green = 0
red_blue_image = img6.copy()
red_blue_image[:, :, 1] = 0

# Green-Blue channel, red = 0
green_blue_image = img6.copy()
green_blue_image[:, :, 0] = 0

# Menampilkan 2 gambar dalam 1 baris
fig, ax = plt.subplots(1, 2)
# fig, ax = plt.subplots(1, 2, figsize=(10, 5))

ax[0].imshow(red_blue_image)
ax[0].set_title("Red-Blue Channel")

ax[1].imshow(green_blue_image)
ax[1].set_title("Green-Blue Channel")

# Mengatur jarak antar subplot (w/h)
plt.subplots_adjust(wspace=0.5)
plt.show()
````

````python
# 3. Tampilkan image baris ke 20-115, kolom 25-120
img7 = img6[20:115, 25:120]
plt.imshow(img7)
````

````python
# 4. Tampilkan image baris ke 5-30, semua kolom, channel Red saja!
red_channel = img6.copy()
red_channel[:, :, 1] = 0
red_channel[:, :, 2] = 0

result_red = red_channel[5:30, :]
plt.imshow(result_red)
````

````python
# 5. Buat 5 kotak berbagai ukuran dan warna yang berbeda dalam satu image
import random

img_boxes = img6.copy()

tinggi = img_boxes.shape[0]
lebar = img_boxes.shape[1]
print("Resolusi image: tinggi x lebar = ", tinggi, " x ", lebar, "\n")

for i in range(5):
    x = random.randint(0, img_boxes.shape[1])
    y = random.randint(0, img_boxes.shape[0])
    width = random.randint(10, 3426)
    height = random.randint(10, 4578)
    color = (random.randint(0, 255), random.randint(0, 255), random.randint(0, 255))
    cv.rectangle(img_boxes, (x, y), (x + width, y + height), color, 50)

plt.imshow(img_boxes)
````
